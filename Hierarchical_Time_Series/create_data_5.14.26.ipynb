```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Define the hierarchy
categories = ['Electronics', 'Home & Garden']
lines = {
    'Electronics': ['TVs', 'Cameras'],
    'Home & Garden': ['Gardening Tools', 'Furniture']
}
models = {
    'TVs': ['55-inch', '65-inch', '75-inch'],
    'Cameras': ['DSLR', 'Mirrorless', 'Compact'],
    'Gardening Tools': ['Rakes', 'Shovels', 'Hoes'],
    'Furniture': ['Tables', 'Chairs', 'Shelves']
}
items = {
    '55-inch': ['SKU001', 'SKU002'],
    '65-inch': ['SKU003', 'SKU004'],
    '75-inch': ['SKU005', 'SKU006'],
    'DSLR': ['SKU007', 'SKU008'],
    'Mirrorless': ['SKU009', 'SKU010'],
    'Compact': ['SKU011', 'SKU012'],
    'Rakes': ['SKU013', 'SKU014'],
    'Shovels': ['SKU015', 'SKU016'],
    'Hoes': ['SKU017', 'SKU018'],
    'Tables': ['SKU019', 'SKU020'],
    'Chairs': ['SKU021', 'SKU022'],
    'Shelves': ['SKU023', 'SKU024']
}

# Generate date range
start_date = pd.Timestamp('2018-01-01')
end_date = pd.Timestamp('2026-04-30')
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Build the full hierarchy
data = []
for category in categories:
    for line in lines[category]:
        for model in models[line]:
            for item in items[model]:
                # Create time series data for each item
                for date in date_range:
                    # Generate realistic demand with trend, seasonality, and noise
                    days_elapsed = (date - start_date).days
                    trend = 0.001 * days_elapsed  # Slight upward trend
                    seasonality = 5 * np.sin(2 * np.pi * date.dayofyear / 365)  # Yearly seasonality
                    noise = np.random.normal(0, 2)  # Random noise
                    
                    # Base demand depends on product type
                    base_demand = 50 if category == 'Electronics' else 30
                    quantity = max(1, int(base_demand + trend + seasonality + noise))
                    
                    data.append({
                        'date': date,
                        'category': category,
                        'line': line,
                        'model': model,
                        'item': item,
                        'quantity': quantity
                    })

# Create DataFrame
df = pd.DataFrame(data)

# Display summary
print("Hierarchical Time Series Dataset Summary")
print("=" * 60)
print(f"Date range: {df['date'].min().date()} to {df['date'].max().date()}")
print(f"Total records: {len(df):,}")
print(f"\nHierarchy levels:")
print(f"  Categories: {df['category'].nunique()}")
print(f"  Lines: {df['line'].nunique()}")
print(f"  Models: {df['model'].nunique()}")
print(f"  Items: {df['item'].nunique()}")
print(f"\nDataFrame shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head(10))
print(f"\nDataFrame info:")
print(df.info())
```

```python
# Export
df.to_csv('hierarchical_time_series_dataset.csv', index=False)
```
